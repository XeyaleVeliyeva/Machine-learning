```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\compo1022\Desktop\DS-128\ML-128\KNN\breast-cancer.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
df
```

```python
df.isna().sum().sort_values(ascending=False).head(20)
```

```python
df.corr()
```

```python
df = df.drop([
    'id',
    'smoothness_se',
    'texture_se',
    'fractal_dimension_mean',
    'symmetry_se',
    'fractal_dimension_se'
], axis=1)
```

```python
from scipy.stats import zscore
import numpy as np
num_cols = df.select_dtypes(include=np.number).columns
for cols in num_cols:
    z_scores = np.abs(zscore(df[cols]))
    outliers = df[z_scores > 3]
    print(cols, ":", len(outliers), "outliers")
df = df[(np.abs(zscore(df[num_cols])) < 3).all(axis=1)]
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('diagnosis',axis=1)
y = df['diagnosis']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
y_pred = knn.predict(X_test)

cm = confusion_matrix(y_test,y_pred)

sns.heatmap(cm,annot=True,cmap='Blues',fmt='d')
```

```python
from sklearn.metrics import classification_report
cl = classification_report(y_test,y_pred)
print(cl)
```

```python
from sklearn.metrics import accuracy_score
accuracies = []
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    accuracies.append(accuracy_score(y_test, knn.predict(X_test)))

import matplotlib.pyplot as plt
plt.plot(range(1, 21), accuracies)
plt.xlabel("k (neighbors)")
plt.ylabel("Accuracy")
plt.title("K seçiminə görə dəqiqlik")
plt.show()
```
